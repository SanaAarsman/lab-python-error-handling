def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))

                    raise ValueError("Invalid quantity! Please enter a non-negative value.")

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
 
def initialize_inventory(prod):
    inventory = {}
    for prod in products:
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {prod}'s available: "))
                if quantity < 0:
                    print()
                    print("Quantity cannot be negative. Please enter a valid quantity.")
                    continue
                inventory[prod] = quantity
                break
            except ValueError:
                print()
                print("Quantity needs to be a integer, please try again")
                print()
    return inventory
            
    
inventory = initialize_inventory(products)
print()
print(f"Inventory contains: {inventory}")        
````

````python
def get_customer_orders():
    customer_orders = set()
    while True:
        try:
            order = input("Do you want to order a product? (Yes/No): ").lower()
            if order not in ["yes", "no"]:
                raise ValueError("Please answer Yes or No.")
            break
        except ValueError as e:
            print(e)
            print()
    if order == "no":
        return customer_orders
    while True:
        try:
            item = input(f"What do you want to order? Choose from {products}: ").lower()
            if item not in products:
                raise ValueError("Invalid product. Please choose something from the list.")
            customer_orders.add(item)
            while True:
                try:
                    again = input("Do you want to add another product? (Yes/No): ").lower()
                    if again not in ["yes", "no"]:
                        raise ValueError("Answer must be Yes or No.")
                    break
                except ValueError as e:
                    print(e)
                    print()
            if again == "no":
                break
        except ValueError as e:
            print(e)
            print()
    return customer_orders


valid_orders = get_customer_orders()
print()
print(f"Customer ordered: {valid_orders}")
print()
````

````python
def calculate_total_price(valid_orders):
    price_products = {}
    if not valid_orders:
        print("No products ordered. Total price is 0.")
        return 0.0
    for prod in valid_orders:
        while True:
            try:
                price = float(input(f"Enter the price of {prod}: "))
                if price < 0:
                    print("Price cannot be negative. Please enter a valid price.")
                    continue
                price_products[prod] = price
                break  
            except ValueError:
                print("Price must be a number. Please try again.")
                print()
    total_price = sum(price_products.values())
    return total_price


total = calculate_total_price(valid_orders)
print()
print(f"Total price of the order: {total}")
````
